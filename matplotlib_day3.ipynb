# Scatter Plot
- plt.scatter()
- scatters data points across x and y axis.
- Scatter plot charts are good for relationships and distributions

```python
#displays function's docstring 
plt.scatter?
```

```python
plt.scatter(x='sepal_length',y='sepal_width', data=df,label='sepal', color='red')
plt.scatter(x='petal_length',y='petal_width',data=df,label='petal', color='blue')
plt.legend()
plt.title("length vs width")
plt.xlabel("length")
plt.ylabel("width")
plt.show()
```

# Bar Graphs
- plt.bar()
- plot a bar graph between heights and labels you provides.
- Bar charts are good for comparisons

```python
year = np.arange(4)
car1 = [40, 17, 28, 12]
car2 = [34, 21, 24, 15]
```

```python
plt.figure(figsize=(8,6))
plt.bar(year,car1,label='BMW', tick_label= [2021,2022,2023,2024])
plt.bar(year,car2,label='Audi')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.show()
```

```python
year
```

```python
plt.figure(figsize=(8,6))
plt.bar(year,car1,width=0.3,label='BMW',tick_label= [2021,2022,2023,2024])
plt.bar(year+0.3,car2,width=0.3,label='Audi')
plt.legend()
plt.xlabel("Years")
plt.ylabel("Sales in Million Dollar")
plt.title("Cars price comparison")
plt.show()
```

# Pie Chart
- plt.pie()
- plot the pie chart with the provides labels and values
- pie charts should be used only for simple compositions never for comparisons or distributions.

```python
subjects = ["Physics", 'Chemistry', 'Maths', "English"]
marks = [78, 45, 90, 65]
```

```python
78/sum(marks)
```

```python
plt.figure(figsize=(10,9))
plt.pie(x= marks,labels=subjects,explode=[0.1,0.1,0.1,0.1],shadow=True, autopct="%0.2f",rotatelabels=True)
plt.legend(subjects)
plt.title("subject wise marks distribution")
plt.show()
```
