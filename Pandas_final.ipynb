```python
import pandas as pd

# Creating a sample DataFrame
data = {'A': [1, 2, 3, 4],
        'B': ['apple', 'banana', 'orange', 'apple'],
        'C': [0.1, 0.2, 0.3, 0.1]}

df = pd.DataFrame(data)
df
# Define the specific value you are interested in
value_to_match = 'apple'

# Retrieve all columns for rows where column 'B' has the specified value
result = df[df['B'] == value_to_match]

# Display the details for the matched rows
print(result)
```

# Merging Dataframes

```python
df1=pd.DataFrame({'Name' : ['Mohit', 'Jatin', 'Prateek', 'Mohit'],
                    'CGPA' :  [ 2 , 4, 5, 3]})
```

```python
df1
```

```python
df2 = pd.DataFrame({'T_Name' : ['Mohit', 'Jatin', 'Prateek', 'Mohit'],
                    'CGPA' :  [ 3 , 6, 8, 9]})
```

```python
df2
```

```python
df1.merge(df2,how='inner')
```

```python
df1.merge(df2, how = 'left')
```

```python
df1.merge(df2, how = 'right')
```

```python
df1.merge(df2, how= 'outer')
```
